## Training using a text dataset via Hugging Face

The following is the `mini-enwiki.yaml` settings, for using a textual dataset via huggingface, with all the comments removed

```.yaml
trainer:
  max_steps: 10
model:
  load_model: ../model/L6-D512-neox-init.pth
  ctx_len: 1024
  lr_init: 3e-4
data:
  data_path: ../datapath/enwiki_10k_world_1024/
  source: "teven/enwiki_10k"
  tokenizer: world
  min_token_size: 64
  max_token_size: -1
  text_rechunk_size: 1024
  text_rechunk_force: true
  custom_text_key: 'text'
  test_split: 0.01
  test_split_shuffle: false
```
